# **Recurrent Neural Networks**

La exploración de las RNN, un tipo de red neuronal artificial, marca un hito importante en la búsqueda de dotar a las máquinas de la capacidad de comprender, interpretar y predecir datos secuenciales.

Una red neuronal recurrente (RNN) es el tipo de red neuronal artificial (ANN) que se utiliza en Siri de Apple y la búsqueda por voz de Google. RNN recuerda entradas pasadas gracias a una memoria interna que es útil para predecir precios de acciones, generar texto, transcripciones y traducción automática.

En la red neuronal tradicional, las entradas y salidas son independientes entre sí, mientras que la salida en RNN depende de elementos anteriores dentro de la secuencia. Las redes recurrentes también comparten parámetros en cada capa de la red. En las redes feedforward, existen diferentes pesos en cada nodo. Mientras que RNN comparte los mismos pesos dentro de cada capa de la red y durante el descenso del gradiente, los pesos y la base se ajustan individualmente para reducir la pérdida.

**¿Qué es un RNN y cómo funciona?**

Comprensión del concepto básico de red neuronal recurrente (RNN)
Una RNN o red neuronal recurrente es un tipo de red neuronal artificial diseñada para reconocer patrones en secuencias de datos. A diferencia de las redes neuronales tradicionales, las RNN utilizan su memoria interna para procesar entradas, lo que las hace ideales para datos secuenciales, como oraciones en el procesamiento del lenguaje natural o datos de series temporales. Esta memoria en RNN permite el almacenamiento de salidas anteriores y su uso para calcular la salida actual, lo que permite a la red mantener una forma de "memoria" a lo largo del tiempo.

**Explorando las características únicas de los RNN en el procesamiento de datos secuenciales**

Las RNN se distinguen de otros tipos de redes neuronales por su capacidad para gestionar y procesar datos secuenciales. La arquitectura de la red incorpora bucles que permiten que la información se transporte a través de las neuronas de un paso al siguiente. Esta estructura aborda la dependencia secuencial en los datos de entrada, lo que hace que los RNN sean excepcionalmente hábiles en tareas que involucran secuencias, como predecir la siguiente palabra en una oración o comprender las dependencias en el lenguaje hablado.

**El papel de las funciones de activación en los RNN**

Las funciones de activación desempeñan un papel crucial en la determinación de la salida de las redes neuronales, incluidas las RNN. Introducen no linealidad en la red, lo que le permite aprender relaciones complejas entre los datos de entrada y salida. En los RNN, las funciones de activación son vitales para la transformación de entradas combinadas con salidas anteriores en la salida actual. Ayudan en la toma de decisiones al determinar qué información pasar, bloquear o modificar antes de pasar al siguiente paso.

### **Cómo funcionan las redes neuronales recurrentes**

En las RNN, la información recorre el bucle, por lo que la salida está determinada por la entrada actual y las entradas recibidas anteriormente.

```python
Image(filename='local/imgs/rnn_1.png', width=400)
```

La capa de entrada X procesa la entrada inicial y la pasa a la capa intermedia A. La capa intermedia consta de múltiples capas ocultas, cada una con sus funciones de activación, pesos y sesgos. Estos parámetros están estandarizados en toda la capa oculta, de modo que en lugar de crear varias capas ocultas, creará una y la recorrerá en bucle.

En lugar de utilizar el backpropagation tradicional, las redes neuronales recurrentes utilizan algoritmos de retropropagación en el tiempo (BPTT) para determinar el gradiente. En la retropropagación, el modelo ajusta el parámetro calculando los errores desde la capa de salida a la de entrada. BPTT suma el error en cada paso de tiempo ya que RNN comparte parámetros en cada capa. 

### **Tipos de redes neuronales recurrentes**

Las redes feedforward tienen una única entrada y salida, mientras que las redes neuronales recurrentes son flexibles ya que se puede cambiar la longitud de las entradas y salidas. Esta flexibilidad permite a los RNN generar música, clasificación de sentimientos y traducción automática.

Hay cuatro tipos de RNN basados en diferentes longitudes de entradas y salidas.

**One-to-one** es una red neuronal simple. Se usa comúnmente para problemas de aprendizaje automático que tienen una única entrada y salida.

**One-to-many** tiene una única entrada y múltiples salidas. Esto se utiliza para generar títulos de imágenes.

**Many-to-one** toma una secuencia de múltiples entradas y predice una única salida. Es popular en la clasificación de sentimientos, donde la entrada es texto y la salida es una categoría.

**Many-to-many** requiere múltiples entradas y salidas. La aplicación más común es la traducción automática.

### **CNN vs RNN**

La red neuronal convolucional (CNN) es una red neuronal de retroalimentación capaz de procesar datos espaciales. Se utiliza comúnmente para aplicaciones de visión por computadora, como la clasificación de imágenes. Las redes neuronales simples son buenas para clasificaciones binarias simples, pero no pueden manejar imágenes con dependencias de píxeles. La arquitectura del modelo CNN consta de capas convolucionales, capas ReLU, capas de agrupación y capas de salida completamente conectadas. 

```python
Image(filename='local/imgs/cnn.png', width=1200)
```

### **Diferencias clave entre CNN y RNN**

* CNN es aplicable para datos escasos como imágenes. RNN es aplicable para series de tiempo y datos secuenciales.

* Mientras entrena el modelo, CNN usa una retropropagación simple y RNN usa retropropagación a lo largo del tiempo para calcular la pérdida.

* RNN no puede tener restricciones en la longitud de las entradas y salidas, pero CNN tiene entradas y salidas finitas.

* CNN tiene una red de retroalimentación y RNN trabaja en bucles para manejar datos secuenciales.

* CNN también se puede utilizar para el procesamiento de vídeos e imágenes. RNN se utiliza principalmente para análisis de voz y texto.

### **Limitaciones de RNN**

Los modelos RNN simples suelen encontrarse con dos problemas importantes. Estos problemas están relacionados con el gradiente, que es la pendiente de la función de pérdida junto con la función de error.

1. El problema del gradiente de desaparición (vanish gradient) ocurre cuando el gradiente se vuelve tan pequeño que la actualización de los parámetros se vuelve insignificante; eventualmente el algoritmo deja de aprender.

2. El problema del gradiente explosivo (exploding gradient) ocurre cuando el gradiente se vuelve demasiado grande, lo que hace que el modelo sea inestable. En este caso, se acumulan gradientes de error mayores y los pesos del modelo se vuelven demasiado grandes. Este problema puede provocar tiempos de entrenamiento más prolongados y un rendimiento deficiente del modelo.

La solución simple a estos problemas es reducir la cantidad de capas ocultas dentro de la red neuronal, lo que reducirá cierta complejidad en las RNN. Estos problemas también se pueden resolver mediante el uso de arquitecturas RNN avanzadas como LSTM y GRU.

### Definamos formalmente las Redes Neuronales Recurrentes

Una visión alternativa de la red

```python
from IPython.display import Image
Image(filename='local/imgs/RNN3.png', width=1200)
```

Según la notación de la figura anterior, la formulación matemática de un Elman RNN con una capa oculta y una capa de salida viene dada por:

$${\bf{a}}^{(t)} = {\bf{b}} + {\bf{V}}{\bf{h}}^{(t-1)} + {\bf{U}}{\bf{x}}^{(t)},\\ {\bf{h}}^{(t)} = \tanh({\bf{a}}^{(t)}), \\ {\bf{o}}^{(t)} = {\bf{c}} + {\bf{W}}{\bf{h}}^{(t)}$$

donde $\bf{V}$ es la matriz de pesos del ciclo de retroalimentación. $\bf{U}$ es la matriz de pesos de las entradas y $\bf{W}$ es la matriz de pesos que conecta el **estado** de la red con la capa de salida. $\bf{b}$ y $\bf{c}$ son los vectores de sesgo para las capas oculta y de salida, respectivamente. La salida de la red ${\bf{y}}^{(t)}$ corresponde a la aplicación de la función de activación a los valores ${\bf{o}}^{(t)}$.

### Backpropagation a través del tiempo (BPTT)

El primer paso para entrenar un RNN es definir la función de pérdida. Consideremos una secuencia de longitud $\tau$, la pérdida de esa secuencia única se puede expresar como:

$$L(\{ {\bf{y}}^{(1)},{\bf{y}}^{(2)},\cdots,{\bf{y}}^{(\tau)} \}, \{ {\bf{\hat{y}}}^{(1)},{\bf{\hat{y}}}^{(2)},\cdots,{\bf{\hat{y}}}^{(\tau)} \}) \\
L = \sum_t L^{(t)}$$

Si, por ejemplo, la función de pérdida para la salida $i$ es $L^{(t)} = -\log \hat{y}_i^{(t)}$, entonces

$$(\nabla_{{\bf{o}}^{(t)}}L)_i = \frac{\partial L}{\partial o_i^{(t)}} = \frac{\partial L}{\partial L^{(t)}} \frac{\partial L^{(t)}}{\partial \hat{y}_i^{(t)}} \frac{\partial \hat{y}_i^{(t)}}{\partial o_i^{(t)}} = \hat{y}_i^{(t)} - 1 \\ \nabla_{{\bf{o}}^{(t)}}L = {\bf{y}}^{(t)} \odot ({\bf{\hat{y}}}^{(t)} - {\bf{1}} )$$

Donde $\odot$ es el producto de Hadamard. Cuando $t=\tau$, ${\bf{h}}^{(\tau)}$ sólo tiene ${\bf{o}}^{(\tau)}$ como descendente, así su gradiente es simple:

$$ \nabla_{{\bf{h}}^{(\tau)}} L = {\bf{W}}^T\nabla_{{\bf{o}}^{(\tau)}}L$$

A partir de estos dos resultados, podemos iterar hacia atrás en el tiempo para propagar gradientes hacia atrás a través del tiempo. De $t=\tau-1$ hasta $\tau = 1$, ${\bf{h}}^{(t)}$ tiene dos descendentes: ${\bf{h}}^{(t+1)}$ y ${\bf{o}}^{(t)}$, por lo tanto su gradiente está dado por:   

$$ \nabla_{{\bf{h}}^{(t)}}L = \left( \frac{\partial {\bf{h}}^{(t+1)}}{\partial {\bf{h}}^{(t)}}\right)^T(\nabla_{{\bf{h}}^{(t+1)}} L) + \left(\frac{\partial {\bf{o}}^{(t)}}{\partial {\bf{h}}^{(t)}}\right)^T(\nabla_{{\bf{o}}^{(t)}}L) \\
\nabla_{{\bf{h}}^{(t)}}L = {\bf{V}}^T\text{diag} \left( 1 - \left( {\bf{h}}^{(t+1)} \right)^2\right)(\nabla_{{\bf{h}}^{(t+1)}} L) + {\bf{W}}^T(\nabla_{{\bf{o}}^{(t)}}L)$$

Con base en los resultados anteriores, el resto de gradientes se pueden calcular como:

$$ \nabla_{\bf{c}} L = \sum_t \left(\frac{\partial {\bf{o}}^{(t)}}{\partial {\bf{c}}}\right)^T \nabla_{{\bf{o}}^{(t)}}L$$

$$ \nabla_{\bf{b}} L = \sum_t \left(\frac{\partial {\bf{h}}^{(t)}}{\partial {\bf{b}}}\right)^T \nabla_{{\bf{h}}^{(t)}}L = \sum_t \text{diag}\left( 1 - \left( {\bf{h}}^{(t)} \right)^2\right)\nabla_{{\bf{h}}^{(t)}} L $$

$$\nabla_{{\bf W}}L = \sum_t \sum_i \left(  \frac{\partial L}{\partial o_i^{(t)}} \right)^T \nabla_{{\bf W}} o_i^{(t)} = \sum_t (\nabla_{{\bf{o}}^{(t)}}L){\bf{h}}^{(t)^T}$$

$$\nabla_{\bf V}L = \sum_t \text{diag}\left( 1 - \left( {\bf{h}}^{(t)} \right)^2\right)(\nabla_{{\bf{h}}^{(t)}} L){\bf{h}}^{(t-1)^T}$$

$$ \nabla_{\bf U}L = \sum_t \text{diag}\left( 1 - \left( {\bf{h}}^{(t)} \right)^2\right)(\nabla_{{\bf{h}}^{(t)}} L){\bf{x}}^{(t)^T} $$

## Encoder-Decoder architecture (sequence-to-sequence)

```python
look_back = 3;
time_ahead = 3
dataset = datasetO.values
dataset = dataset.astype('float32')
# split into train and test sets
train_size = int(len(dataset) * 0.67)
test_size = len(dataset) - train_size
train, test = dataset[0:train_size], dataset[train_size:len(dataset)]
# normalize the dataset
scaler = MinMaxScaler(feature_range=(0, 1))
trainN = scaler.fit_transform(train)
testN = scaler.transform(test)
X_train, y_train = DataPreparationRNN.create_datasetMultipleTimesBackAhead(trainN, n_steps_out=time_ahead, n_steps_in = look_back, overlap = time_ahead)
X_train2, y_train2 = DataPreparationRNN.create_datasetMultipleTimesBackAhead(trainN, n_steps_out=time_ahead, n_steps_in = look_back, overlap = 1)
X_test, y_test = DataPreparationRNN.create_datasetMultipleTimesBackAhead(testN, n_steps_out=time_ahead, n_steps_in = look_back, overlap = time_ahead)
#-----------------------------------------------
model8 = Sequential()
model8.add(SimpleRNN(5,activation = 'relu',input_shape=(look_back,1)))
model8.add(RepeatVector(time_ahead))
model8.add(SimpleRNN(5,activation = 'relu',return_sequences=True))
model8.add(TimeDistributed(Dense(1)))
#-----------------------------------------------------------------------
model8.compile(optimizer='adam',loss='mse')
model8.fit(X_train2.reshape(X_train2.shape[0],look_back,1),y_train2.reshape(y_train2.shape[0],time_ahead,1),epochs=1000, verbose=0)
trainPredict, testPredict = DataPreparationRNN.EstimaRMSE_RNN_MultiStepEncoDeco(model8,X_train,X_test,y_train,y_test,scaler,look_back,time_ahead)
DataPreparationRNN.PintaResultado(dataset,trainPredict,testPredict,look_back)
```
